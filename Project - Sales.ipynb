```python
plt.figure(dpi=98)
plot=merg_df.groupby('Month')[['Customers']].mean().plot(marker='o',color='brown')
plot.set_title('Average Customer per Month')
```

```python
plot2=merg_df.groupby('Day')[['Sales']].mean().plot(marker='*',color='purple')
plot2.set_title('Averayge Sales per Day in a Month')
```

```python
plot3=merg_df.groupby('Day')[['Customers']].mean().plot(marker='v')
plot3.set_title('Average Customers per Day in a Month')
```

```python
plot4=merg_df.groupby('DayOfWeek')[['Customers']].mean().plot(marker='*')
plot4.set_title('Average Customers per Day in a Week')
plot5=merg_df.groupby('DayOfWeek')[['Sales']].mean().plot(marker='+')
plot5.set_title('Average Sales per Day in a Week')
```
